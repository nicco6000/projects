```python
create_new_table='''
CREATE TABLE twitter.trends (
    name varchar(500),
    promoted_content varchar(500),
    query varchar(500),
    tweet_volume int,
    url varchar(500),
    PRIMARY KEY (url)
)
'''
engine.execute(create_new_table)
```

```python
engine.execute('USE twitter')
```

```python
query_template = '''INSERT IGNORE INTO trends(name, 
                                        promoted_content,
                                        query, 
                                        tweet_volume,
                                        url) 
                    VALUES (%s, %s, %s, %s, %s)'''


for entry in top50[0]['trends']:
    print(entry['name'])
    query_parameters = (entry['name'].encode("utf8"), entry['promoted_content'], entry['query'], 
                        entry['tweet_volume'],
                        entry['url'])
    engine.execute(query_template, query_parameters)
    
```

```python
query='''
SELECT name, tweet_volume
FROM trends
ORDER BY tweet_volume DESC
limit 20
'''
df_trends = pd.read_sql(query, con=engine)
df_trends.set_index('name')
```

```python
df_trends.plot(kind='bar', x='name',y='tweet_volume',figsize=(100,100),
                                                           fontsize=100)
```
